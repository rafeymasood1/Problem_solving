```python
class prime:
    def __init__():
        pass
    
    def isprime(number):
        ''' returns True if The number is a prime number and False otherwise | number must be type int'''
        
        for num in range(2,100):
            if number % num == 0 and number != num:
                return False
            
            else:
                continue
            
        return True
    
    
    def is_mersenne_prime(number):
        ''' here the number is 2^(number) -1 , so 3 would return 7 as 7 is prime'''
        number_1 = 2**number -1
        for num in range(2,10000):
            if number_1 % num == 0 and number_1 != num:
                return False
            
            if number_1 == num:
                break
            
            else: 
                continue
        
        return number_1
        
        
        
        
        
        
        
        
```

```python
my_number = prime
```

```python
my_number.isprime(3)
```

```python
for num in range(1,10000000):
    if my_number.is_mersenne_prime(num) != False:
        print(my_number.is_mersenne_prime(num))
```
